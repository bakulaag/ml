```python
import numpy as np
import matplotlib.pyplot as plt


def function(x):
    return (x + 3)**2  

def gradient(x):
    return 2 * (x + 3)  


x = 2  
learning_rate = 0.5
num_iterations = 1000


x_steps = [x]
y_steps = [function(x)]


for i in range(num_iterations):
    x = x - learning_rate * gradient(x)
    
    x_steps.append(x)
    y_steps.append(function(x))


x_vals = np.linspace(-6, 2, 100)
y_vals = function(x_vals)


plt.figure(figsize=(10, 6))

plt.plot(x_vals, y_vals, label='y = (x + 3)²', color='blue')  # Plot the function curve
plt.scatter(x_steps, y_steps, color='red', label='Gradient Descent Steps', zorder=5)  # Plot steps
plt.title('Gradient Descent for Finding Local Minima')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def function(x):
    return (x+3)**2

def gradiant(x):
    return 2 * (x+3)

x = 0
learning_step = 0.1
num_iteration = 100


x_step = [x]
y_step = [function(x)]

for i in range(num_iteration):
    x = x- learning_step * gradiant(x)
    x_step.append(x)
    y_step.append(function(x))


x_vals = np.linspace(-6,2,100)
y_vals = function(x_vals)

# plt.figure(figsize = (10,10))
plt.plot(x_vals, y_vals, label = 'y = (x+3)**2', color ='black')
plt.scatter(x_step,y_step, label='gradiant descent steps', color = 'orange')
plt.title('gradiant descent visualise')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
